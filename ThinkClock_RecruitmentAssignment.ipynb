```python
#Revised code
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.signal import find_peaks
# ( actual data)
data = {
    'Cycle': [i for i in range(1, 252)],  # Creating 251 cycles
    'Voltage': np.random.rand(251) * 4.2,  # voltage values (adjust range as needed)
    'Capacity': np.sin(np.linspace(0, 10, 251))  # Example capacity data
}
df = pd.DataFrame(data)
# Calculating dQ/dV
df['dQ/dV'] = np.gradient(df['Capacity'], df['Voltage'])
# Grouping the data by Cycle
grouped_data = df.groupby('Cycle')
# Creating the  3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Consider adjusting these parameters for peak detection based on the data
peak_threshold = 0.2  # Minimum height for a peak
peak_distance = 50  # Minimum distance between peaks
peaks_to_plot = 2  # Number of peaks to plot per cycle (optional)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') 

plt.savefig('my_plot.png')  # Save the plot as 'my_plot.png'


# Consider adjusting these parameters for peak detection based on your data
peak_threshold = 0.2  # Minimum height for a peak
peak_distance = 50  # Minimum distance between peaks
peaks_to_plot = 2  # Number of peaks to plot per cycle (optional)

for cycle, data in grouped_data:
    peak_indices, _ = find_peaks(data['dQ/dV'], height=peak_threshold, distance=peak_distance)

    # Select and plot a limited number of peaks (optional)
    if peaks_to_plot:
        peak_indices = peak_indices[:peaks_to_plot]

    for i in peak_indices:
        ax.scatter(data['Voltage'][i], data['dQ/dV'][i], cycle, c='red', marker='o', s=50)

ax.set_xlabel('Voltage (V)')
ax.set_ylabel('dQ/dV (Ah/V)')
ax.set_zlabel('Cycle')
ax.set_title('Peak Positions vs. Cycle')

plt.show()

```
